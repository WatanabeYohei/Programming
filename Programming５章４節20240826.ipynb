# （基礎知識）
Ａ　while文：while 文は，繰り返す条件を指定して使用する繰返し制御文であり，次の書式で記述する。

```
while(継続条件)
{
    実行文1
    実行文2
    …
}

例
while(i<100)
{
    i++;
}
```
実行文が１文の場合は，{ }を省略することができる。while 文は，実行文を処理するまえに繰返し条件を判断する前判定構造の制御文である。

---

Ｂ　無限ループ：while 文の条件を1（真）にすると，いつまでも繰返し処理を続ける無限ループを構成できる。while 文を用いた無限ループは，次の書式で記述する。

```
while(1)
{
    実行文1
    実行文2
    …
}

例
while(1)
{
    実行文1
    if(条件)break;
    実行文2
    …
}
```
　while 文を終了し，無限ループから抜けるには，break 文を実行する。上の例では，if 文の条件が真の場合に，break 文が実行され，while 文による無限ループを終了する。

# （基礎知識）
Ａ　do～while文：まえに学んだwhile 文は，実行文を処理するまえに繰返し条件を判断する前判定構造の制御文であった。一方，do～ while 文は，実行文を処理したあとで繰返し条件を判断する後判定構造の制御文である。

do ～ while 文では，はじめから繰返し条件が成立していない場合であっても，実行文は最低一回実行される。このため，少なくとも一回は実行文を実行させたいときに，do ～ while 文を用いる。do ～ while 文は，次の書式で記述する。

```
do
{
    実行文1
    実行文2
    …
}while(条件);

例
do
{
  i++;
}
while(i<100);
```
while（条件）のあとには，「;」が必要であることに注意する。また，実行文が1 文のときは，{ } を省略することができる。

```
09         printf("\n");
10     }
11     return 0;
12 }
```

行07～08　内側のループである。横方向に出力する*の数を決める。実行する文が行08 の1 文のみなので{ }を省略している。

行09　改行を行うためのprintf 関数である。
