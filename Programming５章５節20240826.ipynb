```
01 /* 例題 15 */
02 #include<stdio.h>
03 #define N 10
04 int main(void)
05 {
06     int i, j, work;
07     int shiken[N] = {75, 26, 97, 76, 85, 93, 61, 49, 100, 54};
08     for(i = 0; i 〈 N - 1; i++){
09         for(j = i+1; j 〈 N; j++){
10             if(shiken[i] 〈 shiken[j]){
11                 work = shiken[i];
12                 shiken[i] = shiken[j];
13                 shiken[j] = work;
14             }
15         }
16     }
17     for(i = 0; i 〈 N; i++)
18         printf("shiken[%d] = %d\n", i, shiken[i]);
19     return 0;
20 }
```

行07　整数型配列shiken に成績データを代入する。

行10～14 shiken[i] とshiken[j] を比較して，shiken[j] のほうが大きければ，shiken[i] とshiken[j] を交換する。

# （基礎知識）
Ａ　並べ替えの方法：このプログラムで使用している並べ替えは，2 個の配列要素のデータを次々と比較していく方法である。図8 のように，はじめは変数i = 0 とする。そして，①のように変数j = 1 としてshiken[0]とshiken[1]のデータ75 と26 を比較する。この例では，75 > 26 であるからデータの交換をせずに，そのまま次の②に進む。②では，j = 2 として，
shiken[0]とshiken[2]のデータ75 と97を比較する。この例では，75 < 97 であるからデータの交換を実行して③に進む。以下，同様にして処理を続け，⑨では，j = 9 としてshiken[0] とshiken[9] のデータを比較し，この例では100 > 54 であるため，データの交換をしない。

　このようにして，⑨まで（j < N の間）の処理が終われば，i = 1 として同様の処理をj < N の間繰り返す。これをi <（N - 1），j < N の間（この例ではi < 9，j < 10 の間）続ければ並べ替えが終了する。データが大きい順に並ぶことを降順，小さい順に並ぶことを昇順という。

---

Ｂ　データの交換：たとえば，2 個の配列要素shiken[0] とshiken[2] の値を交換するには，作業用変数workを用いた次のような処理を行う。このようすを図9 に示す。

① shiken[0]の値を作業用変数work に代入する。

② shiken[2]の値をshiken[0]に代入する。

③ workの値をshiken[2]に代入する。
