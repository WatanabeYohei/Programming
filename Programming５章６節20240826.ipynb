#（基礎知識）

Ａ　プロトタイプ宣言：プロトタイプ宣言は，作成した関数の戻り値や引数の書式をコンパイラがチェックするための宣言であり，プリプロセッサ指令の次に記述する。

```
戻り値のデータ型 関数名(引数のデータ型 引数の変数名,…);

例
int fsum(int a, int b);
int fsum（int,int）; //これでも可
```
戻り値や引数がない場合には，voidと記述する。

---
Ｂ　return文：関数は，引数として複数のデータを受け取ることができるが，戻り値は呼び出し側に一つしか返せない。関数は，return文で戻り値を返す。return 文は以下の書式で記述する。

```
return 戻り値;

例
return wa;
```
---
Ｃ　記憶域クラスと有効範囲：変数は宣言の方法によって，使用できる範囲などが異なってくる。コンピュータのメモリがどのように変数を記憶するか，という観点で変数を区別することを**記憶域クラス**という。変数の宣言方法と記憶域クラスの関係を表13 に示す。

image.png

　関数の内部で宣言した変数は，その関数の中だけでしか使用できない。このことは，異なる関数で同じ変数名を使用したとしても，トラブルが生じないことになり，効率的にプログラムをつくることができる。一方，関数の外側で宣言した変数は，すべての関数内で使用できる。変数を宣言する場所と記憶域クラスの関係を図11 に示す。
