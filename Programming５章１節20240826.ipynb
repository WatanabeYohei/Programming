# １　Ｃの特徴

C は，比較的自由な記述ができる高機能な言語であるが，それだけにデバッグやメンテナンスが容易になるように，
読みやすい書き方を心がけることがたいせつである。次に
C の特徴を示す。

---

1.   原則として英小文字でプログラムを書く。
2.   プログラムは関数の集まりとして表される。
3.   行番号がなく自由度の高い書式でプログラムを書ける。
4.   かっこを使って，**関数**や文の範囲を示す簡潔な書き方ができる。
5.   データ型や演算子が数多く用意されている。
6.   コンパイルのまえに行う**前処理機能**がある。
7.   **メモリを直接操作する機能**がある。

※関数：決められた処理を行うプログラム

※前処理機能：pre-processorプリプロセッサという。

※メモリを直接操作する機能：pointerポインタ

```python
%%file p128.c
//C program example
#include<stdio.h>
int main(void)
{
    printf("ELIZABETH\n");
    return 0;
}
```

```python
!gcc p128.c -o p128 && ./p128
```

　→　aaa.cのファイルを作成する（下に内容を記述）

!gcc aaa.c -o bbb && ./bbb
　→　aaa.cをコンパイルしてbbbという名前のオブジェクトファイルを生成、成功したらbbbを実行

---

Ａ　文字：プログラムは原則半角英数字で記述。ただし、出力やコメントは日本語記述可

---

Ｂ　注釈：/* と */ で挟まれた部分はコメントとしてプログラムとしては実行されない。また、 // より右側の記述もコメントとして認識される。

---

Ｃ　プリプロセス：Ｃはコンパイルの前にプリプロセスと呼ばれる前処理を行う機能をもっている。#include指令や#define指令をプリプロセッサ指令という。

Ｄ　ヘッダファイル：標準的な関数あるいは，そのプロトタイプ宣言は，ヘッダファイルとよばれるファイルに格納されている。したがって，プログラム中で使用する関数は，それらが格納されているヘッダファイルを#include指令を用いてプログラム中に取り込むことが必要である。表 1 に，おもなヘッダファイルを示す。


※プロトタイプ宣言とは、関数の名前、戻り値の型、引数の型と数をコンパイラに事前に知らせること
