# （基礎知識）
Ａ　else if文：else if はif 文の条件が偽の場合に，if 文に続いて条件の判定をして，次に実行する文を選択する制御文であり，次の書式で記述する。

```
if(条件１)
{

    条件１が真の場合に実行する文2
    …
}
else if(条件２)
{
    条件１が偽で、条件２が真の場合に実行する文1
    条件１が偽で、条件２が真の場合に実行する文2

    条件１，２ともに偽の場合に実行する文2
    …
}
```
条件が三つ以上ある場合には，else if を続けて記述することができる。else 以下は必要なければ省略することができる。

　また，if 文と同様に実行する文が1 文のみである場合には，{ }を省略することができる。

# （基礎知識）
Ａ　switch文：switch 文は，条件式の値と対応するcase 文を実行する多岐選択制御文であり，次の書式で記述する。

```
switch(式)
{
    case 1:
        式の値が1 の場合に実行する文1
        式の値が1 の場合に実行する文2
        …
        break;
    case 1:
        式の値が2 の場合に実行する文1
        式の値が2 の場合に実行する文2
        …
        break;
    default:
        式の値どれにもはてはまらない場合に実行する文1
        式の値どれにもはてはまらない場合に実行する文2
        …
        break;
}
```
switch文の式の値は整数でなければならない。case 1やcase 2，defaultはラベルとよばれるもので，コロン「: 」をつける。case 文は必要に応じて，さらに続けて記述することができ，default 文は必要がなければ省略できる。
